```python
#Apply it to our example
from sklearn.feature_extraction.text import CountVectorizer


sample_vectorizer = CountVectorizer()
spam_ham = sample_vectorizer.fit_transform(spam_df['text'])
```

```python
print(sample_vectorizer.get_feature_names())
```

```python
print(spam_ham.toarray())
```

```python
spam_ham.shape
```

```python
label = spam_df['spam'].values

```

```python
#Training the Model
from sklearn.naive_bayes import MultinomialNB

NB_class = MultinomialNB()
NB_class.fit(spam_ham, label)
```

```python
test_sample = ['Hey Amir, how is it going','Hi Amir, PLease let me know what have you decided.']
test_sample_coun = sample_vectorizer.transform(test_sample)
```

```python
test_predict = NB_class.predict(test_sample_coun)
```

```python
X = spam_ham
y = label
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_class = MultinomialNB()
NB_class.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_class.predict(X_train)
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

```python
y_predict_test = NB_class.predict(X_test)
y_predict_test
```

```python
cm1 = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm1, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```
